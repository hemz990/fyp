```python
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk
import tkinter.ttk as ttk
import numpy as np
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.utils import load_img, img_to_array
from tensorflow.keras.models import load_model
from keras.preprocessing import image
from tqdm import tqdm
import os
from PIL import Image
from random import random


scratch = load_model("E:/dev/models/scratch_rem.h5", compile=False)
upscale = load_model("E:/dev/models/upscale_div2k.h5", compile=False)
enhance = load_model("E:/dev/models/enhance.h5", compile=False)
scratch.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
upscale.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
enhance.compile(optimizer='adam', loss='mse', metrics=['accuracy'])

img = []
final = []
output_path = "E:/dev/img/output/"

def img_prep(filename):
    img_ = []
    img = load_img(filename, target_size=(128, 128))
    img = img_to_array(img)
    img = img / 255
    img_.append(img)
    img_ = np.array(img)
    img_ = np.reshape(img_, (1, 128, 128, 3))
    print(img_.shape)
    print("processing.....")
    processed_img = proc(img_)
    show(processed_img)
    
    output_file = os.path.join(output_path, "a.jpg")
    save_image(processed_img, output_file)
    
    return processed_img


def proto(path):
    img_ = []
    face_images = sorted(glob.glob(path + '*.jpg'))
    print(len(face_images))
    
    for filename in face_images:
        img = img_prep(filename)
        img_.append(img)
    
    img_ = np.array(img_)
    return img_

def proc(img):
    print(img.shape)
    img_1 = scratch.predict(img)
    # img_2 = upscale.predict(img_1)
    # img_3 = enhance.predict(img_2)
    # print("final : ",img_3.shape)
    img_1 = np.reshape(img_1,(128,128,3))
    # img_3 = np.reshape(img_3,(256,256,3))
    return img_1

def show(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (256,256))  # Resize the image to desired dimensions
    cv2.imshow('Image', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


def save_image(img, output_path):
    # print("Image shape:", img.shape)
    # print("Image data type:", img.dtype)
    try:
        img = (img * 255).clip(0, 255).astype(np.uint8)
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        cv2.imwrite(output_path, img)
        print("Image saved!")
    except Exception as e:
        print(f"Error saving image: {str(e)}")

def start():
    path = 'E:/dev/img/test_img/'
    img = proto(path)
    show(img[0])
    final = proc(img)
    save_image(final[0], 'E:/dev/img/output/3.jpg')
    show(final[0])

# Function to open file explorer and load image
def open_image():
    file_path = filedialog.askopenfilename(filetypes=[("Image Files", "*.jpg;*.jpeg;*.png")])
    if file_path:
        # Load the selected image
        image = Image.open(file_path)
        image.thumbnail((256,256))  # Resize image to fit the window
        img_tk = ImageTk.PhotoImage(image)
        image_label.configure(image=img_tk)
        image_label.image = img_tk  # Save reference to avoid garbage collection
        global img_path
        img_path = file_path
        print(img_path)

# Function to process the image and return the processed image
def process_image():
    global img_path
    if img_path:
        processed_img = img_prep(img_path)
        save_image(processed_img, "E:/dev/img/output/a.jpg")  # Save the processed image
        processed_image = Image.open("E:/dev/img/output/a.jpg")
        processed_image.thumbnail((256, 256))  # Resize image to fit the window
        processed_img_tk = ImageTk.PhotoImage(processed_image)
        processed_image_label.configure(image=processed_img_tk)
        processed_image_label.image = processed_img_tk  # Save reference to avoid garbage collection
        return processed_image  # Return the processed image
    else:
        print("Please select an image to process.")


window = tk.Tk()
window.title("Image Processing GUI")
window.geometry("800x700")
window.configure(bg="#333333")  # Set the background color to dark mode

# Create an image label to display the original image
image_label = tk.Label(window, bg="#ffffff")
image_label.pack(pady=20)

# Create a button to open the file explorer and load the image
upload_button = ttk.Button(window, text="Upload", command=open_image, style="Custom.TButton")
upload_button.pack(pady=10)

# Create a button to process the image
enhance_button = ttk.Button(window, text="Enhance", command=process_image, style="Custom.TButton")
enhance_button.pack(pady=10)

# Create an image label to display the processed image
processed_image_label = tk.Label(window, bg="#ffffff")
processed_image_label.pack(pady=20)

# Create a button to save the processed image
save_button = ttk.Button(window, text="Save", command=lambda: save_image(process_image(), "output.png"), style="Custom.TButton")
save_button.pack(pady=10)

# Create a custom style for the buttons
style = ttk.Style(window)
style.configure(
    "Custom.TButton",
    font=("Helvetica", 12),
    width=15,
    background="#800080",  # Purple button color
    foreground="#ffffff",  # White button font color
)

# Start the main loop
window.mainloop()



```
